#### Backward-Forward Imputation

```python
def impute_time_series(time_series):
    if isinstance(time_series, list):
        time_series = np.array(time_series, dtype=float)
        time_series[time_series == -1] = np.nan
        time_series[time_series == -2] = np.nan
        time_series = pd.Series(time_series).fillna(method='ffill').fillna(method='bfill')
        return time_series.tolist()
    else:
        return time_series


for col in ['HR_Time_Series', 'Respiration_Time_Series', 'Stress_Time_Series']:
    df_timeseries[col] = df_timeseries[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

for col in ['HR_Time_Series', 'Respiration_Time_Series', 'Stress_Time_Series']:
    for index, row in df_timeseries.iterrows():
        time_series = row[col]

        if isinstance(time_series, list):
            imputed_series = impute_time_series(time_series)
            #print(f"Imputed series for index {index}, column '{col}': {imputed_series}")
            df_timeseries.at[index, col] = imputed_series
```

# Data Augmentation + Train-Test Split

```python
# Define columns for augmentation
time_series_columns = ['HR_Time_Series', 'Stress_Time_Series', 'Respiration_Time_Series']
initial_columns = df.columns.difference(time_series_columns + ['final_target']).tolist()
target_column = 'final_target'

# Augmentation functions
def add_noise(series):
    noise = np.random.normal(0, 0.05 * np.std(series), size=len(series))
    return series + noise

def apply_scaling(series):
    scale_factor = np.random.uniform(0.9, 1.1)
    return series * scale_factor
```

```python
#Associated Train-Test Split
df=df_timeseries.iloc[:30]
df_test=df_timeseries.iloc[30:]
```

```python
df_test.shape
```

```python
# Augment the DataFrame
augmented_data = []

# Iterate through each row in the original DataFrame
for _, row in df.iterrows():
    augmented_data.append(row)  # Original row

    # Create separate augmented samples: one with noise, one with scaling, one with both
    for transformation in ['noise', 'scaling', 'both']:
        new_row = row.copy()

        # Apply the selected transformation to each time series column
        for col in time_series_columns:
            if isinstance(row[col], list):
                series = np.array(row[col])

                if transformation == 'noise':
                    new_row[col] = add_noise(series)
                elif transformation == 'scaling':
                    new_row[col] = apply_scaling(series)
                elif transformation == 'both':
                    new_row[col] = add_noise(apply_scaling(series))
            else:
                new_row[col] = row[col]  # Leave non-list values unchanged

        augmented_data.append(new_row)

# Create the final augmented DataFrame
augmented_df = pd.DataFrame(augmented_data)
```

```python
augmented_df.shape
```

# Ragged Tensor Application

```python
time_series_columns = ['HR_Time_Series', 'Stress_Time_Series', 'Respiration_Time_Series']

#df_train
X_train_ragged_classification = [
    tf.ragged.constant([np.array(ast.literal_eval(row[col]), dtype=float) if isinstance(row[col], str)
                        else np.array(row[col], dtype=float) for col in time_series_columns], dtype=tf.float32)
    for _, row in augmented_df.iterrows()
]

Y_train_classification = augmented_df['Classification_target'].values

#df_test
X_test_ragged_classification = [
    tf.ragged.constant([np.array(ast.literal_eval(row[col]), dtype=float) if isinstance(row[col], str)
                        else np.array(row[col], dtype=float) for col in time_series_columns], dtype=tf.float32)
    for _, row in df_test.iterrows()
]
Y_test_classification = df_test['Classification_target'].values

# Combine the list of RaggedTensors into a single RaggedTensor
X_train_ragged_combined = tf.ragged.stack(X_train_ragged_classification)
X_test_ragged_combined = tf.ragged.stack(X_test_ragged_classification)

print("X_train_ragged shape:", X_train_ragged_combined.shape)
print("Y_train shape:", Y_train_classification.shape)
print("X_test_ragged shape:", X_test_ragged_combined.shape)
print("Y_test shape:", Y_test_classification.shape)
```

# Model Example LSTM, GRU, Simple RNN

```python
import os
import tensorflow as tf
import tensorflow.keras.layers as tfkl
from tensorflow.keras.callbacks import TensorBoard
from datetime import datetime
```

```python
MAX_SEQUENCE_LENGTH = 270
def create_masked_rnn_model(input_shape):
    input_layer = tfkl.Input(shape=input_shape, dtype=tf.float32, ragged=True)

    # Convert RaggedTensor to Dense with specified maximum length
    x = tfkl.Lambda(lambda x: x.to_tensor(shape=(None, MAX_SEQUENCE_LENGTH, 4)), name='ragged_to_dense')(input_layer)

    # Masking
    x = tfkl.Masking(mask_value=0.0)(x)

    x = tfkl.Bidirectional(tfkl.LSTM(8, return_sequences=True))(x)
    x = tfkl.Dropout(0.3)(x)
    x = tfkl.Bidirectional(tfkl.GRU(4, return_sequences=True))(x)
    x = tfkl.Dropout(0.3)(x)
    x = tfkl.Bidirectional(tfkl.SimpleRNN(2))(x)
    x = tfkl.Dropout(0.3)(x)

    # Output layer for binary classification
    output_layer = tfkl.Dense(1, activation='sigmoid')(x)

    model = tf.keras.Model(inputs=input_layer, outputs=output_layer)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

    input_shape = (None, 4)
    model = create_masked_rnn_model(input_shape)
    model.summary()

log_dir = os.path.join("logs", "fit", datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
input_shape = (None, 4)
model = create_masked_rnn_model(input_shape)

# Configura directory di TensorBoard
log_dir = os.path.join("logs", "fit", datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

history = model.fit(
    X_train_ragged_combined, Y_train_classification,
    epochs=5,
    batch_size=4,
    callbacks=[tensorboard_callback],
    verbose=1
)

test_loss, test_accuracy = model.evaluate(X_test_ragged_combined, Y_test_classification, verbose=1)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")

#print(f"Esegui il seguente comando per visualizzare il training su TensorBoard:\ntensorboard --logdir {log_dir}")
```
